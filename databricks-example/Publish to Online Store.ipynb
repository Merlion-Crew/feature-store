# Create & Publish To Online Store

Before moving forward, please ensure you have an instance of an Azure SQL Database. Instructions on how to create and access one can be found here - https://docs.microsoft.com/en-us/azure/azure-sql/database/single-database-create-quickstart?view=azuresql&tabs=azure-portal

```
from databricks import feature_store

fs = feature_store.FeatureStoreClient()
```

```
import datetime
from databricks.feature_store.online_store_spec.azure_sql_server_online_store_spec import AzureSqlServerSpec

online_store = AzureSqlServerSpec(
        hostname='<YOUR-DB-HOSTNAME>',
        port=1433,
        user='<YOUR-DB-USER>',
        password='<YOUR-DB-PASS>',  
        database_name='<YOUR-DB-NAME>',
        table_name='<YOUR-DB-TABLE-NAME>'
)

fs.publish_table(
  name='feature_store_home_credit_bureau_data.installment_payments_features',
  online_store=online_store,
  mode='merge'
)
```
