```
# Populate Environment Variables
import os
os.environ['KAGGLE_USERNAME'] = dbutils.secrets.get(scope="default", key="KAGGLE-USERNAME") # Create scope by navigating to <databricks-url>#secrets/CreateScope
os.environ['KAGGLE_KEY'] = dbutils.secrets.get(scope="default", key="KAGGLE-KEY")
os.environ['KAGGLE_COMPETITION'] = "home-credit-default-risk" # Change this to download data from another competition
```

```
from kaggle.api.kaggle_api_extended import KaggleApi
api = KaggleApi()
api.authenticate()
```

```
# Download Kaggle data to Databricks file system

import zipfile

# Create paths in file system
zip_path = "/databricks/driver/data/zipped-risk-data/"
extracted_path = "/databricks/driver/data/extracted-risk-data/"
dbfs_path = "dbfs:/data/credit-default-risk-data/"

dbutils.fs.mkdirs(zip_path)
dbutils.fs.mkdirs(extracted_path)
dbutils.fs.mkdirs(dbfs_path)

# Download data from Kaggle
api.competition_download_files(os.getenv('KAGGLE_COMPETITION'), path=zip_path)

# Unzip all files into path 
with zipfile.ZipFile(zip_path + os.getenv('KAGGLE_COMPETITION') + ".zip") as z:
    z.extractall(extracted_path)
    print("Extracted all")

# Copy data over to Databricks file system
dbutils.fs.cp(extracted_path, dbfs_path, recurse=True)
dbutils.fs.ls(dbfs_path)
```
